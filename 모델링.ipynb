```python
data_weather = pd.read_excel('weather_x.xlsx')

data_weather
```

```python
data_movie = pd.read_excel('movie_y.xlsx')
data_movie
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_weather,data_movie['customer'],random_state=0 )
```

```python
print("X_train의 크기: {}".format(X_train.shape))
print("y_train의 크기: {}".format(y_train.shape))
print("X_test의 크기: {}".format(X_test.shape))
print("y_test의 크기: {}".format(y_test.shape))
```

```python
from sklearn.neighbors import KNeighborsRegressor

reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)
```

```python
print("테스트 세트 예측:\n")
reg.predict(X_test)
```

```python
print("테스트 세트 R^2:")
reg.score(X_test, y_test)
```
